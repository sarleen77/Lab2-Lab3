### Dataset: Cardiovascular Risk Dataset


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```
# Load Dataset
df = pd.read_csv('cardiovascular_risk_dataset.csv')
df.head()
```

```
# Data Types
df.dtypes
```

```
# Check Missing Values
df.isna().sum()
```

```
# Check Duplicates
df.duplicated().sum()
```

```
# Dataset Shape
df.shape
```

```
# Descriptive Statistics
df.describe(include='all')
```

## Univariate Analysis


```
df.hist(figsize=(12,10))
plt.show()
```

## Correlation Matrix


```
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

## Target Variable Distribution


```
# Assuming target column is named 'target' or similar
if 'target' in df.columns:
    sns.countplot(x='target', data=df)
    plt.show()
else:
    print('Please check the target column name in the dataset.')
```
